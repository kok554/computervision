```python
#pandas 라이브러리를 불러옴
import pandas as pd

# 깃허브에 준비된 데이터를 가져옵니다
!git clone https://github.com/taehojo/data.git

# 광물 데이터를 불러옵니다
df = pd.read_csv('./data/sonar3.csv',header=None)

df.head() # 첫 다섯줄을 봅니다
```

```python
df[60].value_counts()
```

```python
X = df.iloc[:,0:60]
y = df.iloc[:,60]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# 모델을 설정합니다.
model = Sequential()
model.add(Dense(24, input_dim= 60, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# 모델을 컴파일합니다
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# 모델을 실행
history = model.fit(X, y, epochs=200, batch_size=10)
```

## 과적합
### 모델의 학습 데이터셋 안에서는 일정 수준 이상의 예측 정확도를 보이지만, 새로운 데이터에 적용하면 잘 맞지 않는 것을 의미

```python
from sklearn.model_selection import train_test_split

# 학습셋과 테스트셋을 구분합니다.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True)
```

```python
score = model.evaluate(X_test, y_test)
print('Test accuracy:',score[1])
```

### 딥러닝, 머신 러닝의 목표는 학습셋에서만 잘 작동하는 모델을 만드는 것이 아님

### 새로운 데이터에 대해 높은 정확도를 안정되게 보여 주는 모델읆 만드는 것이 목표

### 모델 성능의 향상을 위한 방법에는 크게 데이터를 보강하는 방법과 알고리즘을 하는 방법

### 딥러닝의 경우 샘플 수가 많을수록 성능이 좋아짐을 보여 줌

## K겹 교차검증
### 데이터셋을 여러 개로 나누어 하나씩 테스트셋으로 사용하고 나머지를 모두 합해서 학습셋으로 사용하는 방법
### 100%를 학습셋으로 사용할 수 있고, 또 동시에 테스트셋으로도 사용할 수 있음

### 데이터를 원하는 수 만큼 나누어 각각 학습셋과 테스트셋으로 사용되게 하는 함수는 사이킷런 라이브러리의 KFold()함수

```python
from sklearn.model_selection import KFold
k = 5
kfold = KFold(n_splits=k, shuffle=True)
acc_score = []

for train_index, test_index in kfold.split(X):
    X_train, X_test = X.iloc[train_index,:], X.iloc[test_index,:]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
```

```python
accuracy = model.evaluate(X_test, y_test)
acc_score.append(accuracy)
```
